# Discrete random variables

## A discrete r.v.

```r
x<-c(1,2,3,4)
d<-c(0.2, 0.3, 0.4, 0.1)
```

#### Probability mass function (pmf)

```r
plotmass(x,d)
```

#### A visualization of the probability distribution as a set of probability mass points on the line

```r
plotmasspoints(x,d)
```

#### The cumulative distribution function (cdf) or probability distribution function

```r
plotdist(x,d)
```

#### Joint plot of pmf and cdf

```r
plotmassdist(x,d)
```

## Constant (degenerate) r.v.

#### Vector of values, $x$, and  vector of probabilities, $d$

Both have a single element each

```r
x<-2
d<-1
```

#### Joint plot of pmf and cdf

```r
plotmassdist(x,d)
```

## Hypergeometric r.v.

```r
# Results of drawing k=4 balls without replacement from an urn containing m=6 balls, n=4 of which are white 
#(and m-n=2 are black). The hypergeometric random variable $X$ is the number of white balls obtained.
m<-6
n<-4
k<-4
# Vector of values
x<-0:4
# Vector of probabilities
d<-dhyper(x, m, n, k)
```

#### Joint plot of pmf and cdf

```r
plotmassdist(x,d)
```

## Binomial r.v.

```r
# Binomial B(6,0.4) r.v.
n<-6
p<-0.4
x<-0:n
```

```r
d<-dbinom(x, size=n,prob=p)
plotmassdist(x,d)
```

# Continuous random variables

## Uniform r.v. on a (compact) interval

#### Probability density function (pdf) and cumulative distribution function (cdf)

The shape of this density gives the distribution its name as "rectangular distribution"

```r
# Uniform distribution on [0,2]
f<-function(x){dunif(x,0,2)}
F<-function(x){punif(x,0,2)}
```

#### Joint plot of pdf and cdf

```r
xmin<--1
xmax<-3
plotdensdist(f,F,xmin,xmax)
```

## Exponential r.v.

#### Probability density function (pdf) and cumulative distribution function (cdf)

```r
# Exponential distribution with lambda=2
f<-function(x){dexp(x,rate=2)}
F<-function(x){pexp(x,rate=2)}
```

#### Joint plot of pdf and cdf

```r
xmin<--1
xmax<-6
plotdensdist(f,F,xmin,xmax)
```
