## Exercise 2

Write and run a slice sampler for a univariate $\operatorname{Beta}(3,7)$ pdf. 
Display the traceplot, ACF and histogram of its output.

```r
# Parameters for the target Beta pdf
alpha<-3
beta<-7
f<-function(x){dbeta(x,shape1=alpha,shape2=beta)}
# (a,b) is the interval where f>0
# x0 is the starting point (must be in the (a,b) interval)
# n is the number of steps 
a<-0
b<-1
x0<-0.68
n<-10000
# Main loop: 
#
# At step i, a1 and b1 are the two points where the horizontal line at y cuts the function
#
simple.slice.sampler<-function(f,n,a,b,x0){
    x<-rep(0,(n+1))
    x[1]<-x0
    for (i in 1:n){
        y<-runif(1,min=0,max=f(x[i]))
        fy<-function(x){f(x)-y}
        a1<-uniroot(fy,interval=c(a,x[i]))$root
        b1<-uniroot(fy,interval=c(x[i],b))$root
        x[i+1]<-runif(1,min=a1,max=b1)
        }
    return(x)
    }
# 
x<-simple.slice.sampler(f,n,a,b,x0)
```

```r
# Display results: traceplot ACF and histogram
old.par<-par(mfrow=c(2,1))
plot(1:length(x),x,type="l",col="cyan",xlab="",main="Traceplot")
acf(x,col="cyan")
par(old.par)
```

```r
require(MASS)
truehist(x)
u<-seq(a,b,length=600)
lines(u,f(u),lwd=2.5,col="blue")
```
