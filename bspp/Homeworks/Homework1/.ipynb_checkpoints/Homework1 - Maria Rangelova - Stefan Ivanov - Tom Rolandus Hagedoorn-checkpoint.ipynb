\begin{equation} \label{eq1}

**$(i)$ Find the pmf of $X_2$ conditional to a given value of $X_1$.**

**1.3 Define the r.v. $Y$ = *Total number of tosses.*. Obtain the cdf of Y.**

Let $H^k_j, k = 1, 2, ..., n$ be a random variable indicating that coin $k$ is still not removed after $j$ tosses, i.e.

\begin{equation} \label{eq7}
H^k_j = 
\begin{cases}
    1, & \mbox{if coin k was tails in all j tosses} \\
    0, & \mbox{otherwise}
\end{cases}
\end{equation}

\begin{equation} \label{eq8}
 P(H^k_j=1) = \frac{1}{2^j}
\end{equation}

\begin{equation} \label{eq9}
 P(H^k_j=0) = 1 - \frac{1}{2^j}
\end{equation}


Let $H_j$ be a random variable indicating that there are still coins remaining after $j$ tosses, i.e.

\begin{equation} \label{eq10}
H_j = 
\begin{cases}
    1, & \mbox{if at least one } H^k_j=1, k=1,2,...,n \\
    0, & \mbox{if } H^1_j=H^2_j=...=H^n_j=0
\end{cases}
\end{equation}

Since $H^k_j, k = 1, 2, ..., n$ are all independent random variables,

\begin{equation} \label{eq11}
\begin{split}
P(H_j=0) & = P(H^1_j=0, H^2_j=0, ..., H^n_j=0) \\
 & = \prod_{k=1}^{n} P(H^k_j=0) \\
 & = \prod_{k=1}^{n} 1 - \frac{1}{2^j} \\
 & = (1 - \frac{1}{2^j})^n \\
\end{split}
\end{equation}

Note that the cumulative distribution of $Y$ ($F_Y(j) = P(Y \leq j))$ is the probability of not having any remaining coins after $j$ tosses which is exactly $P(H_j=0)$ and therefore we get

\begin{equation} \label{eq12}
\begin{split}
F_Y(j) & = P(Y \leq j) \\
 & = P(H_j=0) \\
 & = (1 - \frac{1}{2^j})^n \\
\end{split}
\end{equation}

The purpose of this exercise is to simulate a normal distribution $\sim\operatorname{N}(0,1)$ from a Laplace distribution $\operatorname{Laplace}(\alpha)$. We follow the guidelines from the Simulation 3 Notebook to solve it.

Assume we want to generate RN following the $\operatorname{N}(0,1)$ distribution, whose pdf is:

$$
	f(x)=\dfrac{1}{\sqrt{2\,\pi}}\,\exp\left\{-\dfrac{1}{2}\,x^{2}\right\}.
$$

Our candidate pdf is in the $\operatorname{Laplace}(\alpha)$ family:

$$
	g(x|\alpha)=\dfrac{\alpha}{2}\,\exp\left\{-\alpha\,x\right\}.
$$

```python
#install.packages("extraDistr",dependencies=TRUE,repos="https://cloud.r-project.org")
require(extraDistr)
x<-seq(-6,6,length=500)
alpha<-1
plot(x,dlaplace(x,mu=0,sigma=1),type="l",lwd=2.0,col="DarkRed",xlab="x",ylab="Density",main="Normal and Laplace pdf's")
lines(x,dnorm(x),lwd=2.0,col="DarkGreen")
```

#### cdf  for a $\operatorname{Laplace}(\alpha)$

$$
	G(x|\alpha)=\left\{\begin{array}{lll}
	\dfrac{1}{2}\,\exp(\alpha\,x),	&\text{ if }	&x<0,\\
	1-\dfrac{1}{2}\,\exp(-\alpha\,x),  &\text{ if }	&0\leq x.
	\end{array}\right.
$$

#### Quantile function for a $\operatorname{Laplace}(\alpha)$

$$
	Q(y|\alpha)=\left\{\begin{array}{lll}
	\hphantom{-}\dfrac{1}{\alpha}\,\log(2\,y),	&\text{ if }	&y\in(0,0.5),\\
	-\dfrac{1}{\alpha}\,\log(2(1-y)),  &\text{ if }	&y\in(0.5,1).
	\end{array}\right.
$$

The next piece of code generates $\operatorname{Laplace}(\alpha)$-distributed random numbers by the inverse cdf method. We have compared this to the function from R to confirm results:

```python
require(MASS)

# this is the inverse as given above, we use it to generate random numbers by the Laplace distribution
my.rlaplace<-function(x,a){
    u<-runif(x)
    l<-ifelse(u<0.5,(log(2*u))/a,(-log(2-2*u))/a)
    return(l)
}

x<-1000
a<-1
x1<-rlaplace(x, 0, 1)
x2<-my.rlaplace(x, a)
old.par<-par(mfrow=c(1,2))
u<-seq(-5,5,length=1000)
truehist(x1,main="Generated with rlaplace")
lines(u,dlaplace(u,0,1),lwd=2.5,col="blue")
truehist(x2,main="Generated with my.rlaplace")
lines(u,dlaplace(u,0,1),lwd=2.5,col="blue")
```

#### Find a suitable $\alpha$ and the proportion $M$ for the acceptance-rejection algorithm

Compute the quotient:

$$
	\dfrac{f(x)}{g(x)}=\sqrt{\dfrac{2}{\pi}}\,\dfrac{1}{\alpha}\,
	\exp\left\{-\dfrac{1}{2}\,x^{2}+\alpha\,|x|\right\}
$$

For $x>0$, the exponent is $>0$ when $x\in(0,2\,\alpha)$, with maximum $\alpha^{2}/2$ at $x=\alpha$. 

Symmetrically for $x<0$.
	
Substituting, we get a function of $\alpha$:

$$
	K(\alpha)=\dfrac{f(\alpha)}{g(\alpha)}
	=\sqrt{\dfrac{2}{\pi}}\,\dfrac{1}{\alpha}\,e^{\alpha^{2}/2}.
$$

The maximum of $K(\alpha)$ is attained for $\alpha=1$. Then,

$$
	K(\alpha)\leq K_{\max}=\sqrt{\dfrac{2\,e}{\pi}}=1.315489.
$$

We take this value as $M$.

Next, we generate a sequence of $\operatorname{N}(0,1)$-distributed random numbers with the acceptance-rejection algorithm, using a $\operatorname{Laplace}(\alpha=1)$ candidate pdf and proportionality constant $M=1.315489$. We have used the accept-reject function as given to us in the notebook from class.

We have plotted a histogram of the resulting numbers for n = 100, 1000, 10000 and superimposed to it the $\operatorname{N}(0,1)$ pdf.

We have printed the acceptance rate from the theoretical formula 1/c and the observed one for each n as produced by the Accept.Reject function.

```python
Accept.Reject<-function(f,g,rg,c,n=1.0e4){
    z<-rep(0,n) 
    N<-0
    for (i in 1:n){
        accept<-FALSE
        while (accept == FALSE){
            u <- runif(1)
            v <- rg(1)
            N<-N+1
            if (c*u <= f(v)/g(v)){
                z[i] <- v
                accept <- TRUE
                }
            }
        }
    Proportion<-n/N
    return(list(z=z,Proportion=Proportion))
    }
```

```python
require(MASS)

# f is the normal distribution
f<-function(x){
    return(dnorm(x,0,1))
}

# g is laplace
g<-function(x){
    return(dlaplace(x,0,1))
}

# rg needs to generate random numbers from g - we can use the method we already wrote my.rlaplace
rg<-function(x){
    return(my.rlaplace(x,1))
}

# we set the parameter c, as computed above
c<-1.315489
```

```python
options(repr.plot.width=8, repr.plot.height=4)
par(mfrow=c(1,3))

print(paste0("Theoretical acceptance rate: ", 1/c))
for (i in 2:4) {
    l = Accept.Reject(f,g,rg,c,n=10**i)
    u<-seq(-5,5,length=10**i)
    p<-f(u)
    truehist(l$z)
    lines(u,p,lwd=2.5,col="blue")
    print(paste0("n=",10**i))
    print(paste0("Observed acceptance rate: ", l$Proportion))
}

```

Obviously, the resulting shape of the histogram is that of the normal distribution. The observed acceptance rate is also as expected - very close to the theoretical one 1/c.

We find $\alpha_0$ and $\beta_0$ such that $\mbox{Beta}(\alpha_0,\beta_0)$ has the same expectation and variance than the piecewise linear nonconjugate prior by doing the following steps:

* Analytically we find the mean and standard deviation of the function $h$
* Then we use the formula from case study 2 to get $\alpha_0$ and $\beta_0$ from the mean and standard deviation:

$$\begin{aligned}
    \alpha_0  &=\mu\cdot\left[\dfrac{\mu\,(1-\mu)}{\sigma^{2}}-1\right],\\[0.4cm]
    \beta_0   &=(1-\mu)\cdot\left[\dfrac{\mu\,(1-\mu)}{\sigma^{2}}-1\right].\end{aligned}
$$

We then add these values to the table in order to be more efficient in our coding.

```python
# find a0 and b0 such that the distribution has the same mean and std as the piecewise linear function

# analytically we can find the mean and std for the function
h_m <- 0.493
h_s <- 0.21

# then we simply use the formula as given in case study 2 to get alpha and beta from these
h_s2<-h_s^2
G<-(h_m*(1-h_m))/h_s2-1
alpha0<-h_m*G
beta0<-(1-h_m)*G

# add these values to the list that we are going to evaluate
alphas[8] <- alpha0
betas[8] <- beta0
table_values1 <- alphas /(alphas + betas)
table_values2 <- alphas + betas
```

The posterior of a pair formed by a Bernoulli likelihood and a Beta prior, which is a conjugate pair, is given by $\textrm{Beta}(\alpha + n1, \beta + n - n1)$ where $n$ is the sample size and $n_1$ the number of samples of the target category. Following the Bernoulli model with a conjugate prior, the prior predictive and the posterior predictive pdf's are given by the following Beta-Binoamials: $\textrm{BetaBin}(n,\alpha,\beta)$ and $\textrm{BetaBin}(n,\alpha + n1, \beta + n - n1)$.

```python
mode <- function(v) {
   uniqv <- unique(v)
   uniqv[which.max(tabulate(match(v, uniqv)))]
}

N<-1000
n<-980
n1<-437

post_pred_median <- rep(0,8)
E <- rep(0,8)
Q_lower <- rep(0,8)
Q_upper <- rep(0,8)
MAP <- rep(0,8)

for (i in 1:8) {
    
    a = alphas[i]
    b = betas[i]
    
    # posterior - beta with a+n1, b+n-n1
    posterior <- rbeta(N,a+n1,b+n-n1)

    # prior predictor - beta-binomial with a,b
    prior_predictor <- rbbinom(N, n, a, b)

    # posterior predictor - beta-binomial with a+n1,b+n-n1
    posterior_predictor <- rbbinom(N, n, a+n1, b+n-n1)
    
    # posterior expectation of theta - we can check the formula from the lectures
    # against what the data shows to confirm, and indeed they are the same
    # (with some rounding error)
    m <- median(sort(posterior))
    e <- (a+n1)/(a+b+n)
    
    # 95% credible interval
    q <- quantile(posterior, probs=c(0.025,0.975))
    
    # MAP for theta, which is essentialy the mode of the posterior
    map <- mode(round(posterior,3))
    
    # save these so we can create the table
    post_pred_median[i] <- m
    E[i] <- e
    Q_lower[i] <- q[1]
    Q_upper[i] <- q[2]
    MAP[i] <- map
    

    
}


#CREATING THE TABLE 2.1
table_2_1 <- data.frame(alphas = round(alphas,2), betas = round(betas,2), p = table_values1, t = table_values2, 
                        E = round(E,3) , Q_lower = round(Q_lower,3), 
                       Q_upper = round(Q_upper,3), MAP = round(MAP,3))
table_2_1 <-  dplyr::rename(table_2_1,'$\\frac{\\alpha}{\\alpha + \\beta}$' = p,
                         '$\\alpha+\\beta$' = t,
                         'Posterior  median of $\\theta$' = E,
                          'Lower bound of 95% interval' = Q_lower,
                        'Upper bound of 95% interval' = Q_upper,
                           '$\\alpha$' = alphas,
                           '$\\beta$' = betas)

for (i in 1:nrow(table_2_1)){
    table_2_1$'95% posterior interval for $\\theta$' [i] <- paste('[', paste(toString(table_2_1$'Lower bound of 95% interval'[i]),
                               toString(table_2_1$'Upper bound of 95% interval'[i]),sep = ", "),
                               ']', sep = "")
}

table_2_1$'Lower bound of 95% interval' <- NULL
table_2_1$'Upper bound of 95% interval' <- NULL

table_2_1 <- table_2_1[,c(1,2,3,4,5,7,6)]
table_2_1

```

### 3.2

* Simulate the prior predictive pmf with the nonconjugate piecewise linear pdf. Do it in both possible ways: (1) Sample from $h$ by the inverse transformation method and (2) Sample from its discretization. 


* Summarize the resulting simulated sample: plot the empirical pmf, compute median, mode, expectation, standard deviation. Compare the results with those obtained from the conjugate prior $\mbox{Beta}(\alpha_0,\beta_0)$. 

We know h as defined in the notebook from class, so we simply imlpement it in code:
$$
    h(t)=\left\{
    \begin{array}{lcl}
         a,        &\text{ for }  &0<t<m-c \quad\text{and } m+c<t<1,\\
         A+b\cdot t/c,  &\text{ for }  &t\in [m-c,c],\\ 
         B-b\cdot t/c   &\text{ for }  &t\in [c,m+c],\\ 
    \end{array}
    \right.,\quad t \in[0,1]
$$

For the inverse transformation method, we also need to compute the inverse:
$$
Q(y) = \left\{
        \begin{array}{lcl}
        2y, &\mbox{ if } & y\in(0,0.1925),\\
        0.375 + 0.02 \sqrt{100y-19}, &\mbox{ if } & y\in[0.1925,0.4925],\\
        0.595 - 0.0141421 \sqrt{159-200y}, &\mbox{ if } & y\in[0.4925,0.7925],\\
        2y - 1, &\mbox{ if } & y\in(0.7925,1).
        \end{array}
        \right.
$$

Here follows the implementations of the two methods:

```python
inv<-function(n){
    u<-runif(n)
    l<-ifelse( u<0.1925, 2*u, 
               ifelse( u<0.4925, 0.375 + 0.02*sqrt(100*u-19), 
                       ifelse( u<0.7925, 0.595 - 0.0141421*sqrt(159-200*u), 2*u-1 )))
    return(l)
}

N = 10000
u<-seq(0,1,length=N)
inv_pr_pred<-rbinom(N, n, inv(u))
hist(inv_pr_pred,main="Generated with the inverse method",nclass = 200)
```

```python
h<-function(t){
    
    m<-0.485
    c<-0.100
    p.tails<-0.4  
    a<-0.5
    b<-5.0
    A<-a+b-(b*m)/c
    B<-b+a+(b*m)/c
    
    if ((0<=t && t<m-c) || (m+c<t && t<=1)) {
        return(a)
    } else if (m-c<=t && t<=m) {
        return(A+b*t/c)
    } else if (m<=t && t<=m+c) {
        return(B-b*t/c)
    }
    return(0)
    
}


t<-seq(0,1,length=N)
t<-t[-N]
probs<-lapply(t,function(x){h(x)/N})
discretetization_sampling <- sample(t, size = N, replace = TRUE, prob = probs)
disc_pr_pred <- rbinom(N, n, discretetization_sampling)
hist(disc_pr_pred,main="Sampling from the discretization",nclass = 200)
```

```python
message("From the inverse transformation we obtain the median: ", round(median(inv_pr_pred),1))
message("From the inverse transformation we obtain the mode: ", round(mode(inv_pr_pred),1))
message("From the inverse transformation we obtain the expectation: ", round(mean(inv_pr_pred),1))
message("From the inverse transformation we obtain the standard deviation: ", round(sd(inv_pr_pred),1),"\n")

message("From the discretization we obtain the median: ", round(median(disc_pr_pred),1))
message("From the discretization we obtain the mode: ", round(mode(disc_pr_pred),1))
message("From the discretization we obtain the expectation: ", round(mean(disc_pr_pred),1))
message("From the discretization we obtain the standard deviation: ", round(sd(disc_pr_pred),1),"\n")
```

### 3.3

* Simulate the posterior predictive pmf (given the observed data of 437 female births from 980 placenta previa births) with the nonconjugate piecewise linear pdf. Sample from the the discretization of the posterior pdf as in Exercise 2.


* Summarize the resulting simulated sample: plot the empirical pmf, compute median, mode, expectation, standard deviation. Compare the results with those obtained from the conjugate prior $\mbox{Beta}(\alpha_0,\beta_0)$.

```python
n<-980
x<-437
N1<-1000
t<-seq(0,1,length=N)
t<-t[-N]
N = 100000

Likelihood <- function(x,theta,n){theta^x*(1-theta)^(n-x)}
nnpost <- Likelihood(x,t,n)*h(t)
dpost<-nnpost/sum(nnpost)
probs <- 1/N1*dpost
discrete_posterior = sample(t, size = N, replace = TRUE, prob = probs)
discrete_post_predictive<-rbinom(N, n, discrete_posterior)

truehist(discrete_post_predictive, xlab = "t_i", ylab = "h(t_i)",
         main = "Sampling from the discretized posterior")
```

```python
message("From the discretization we obtain the median: ", round(median(discrete_post_predictive),3))
message("From the discretization we obtain the mode: ", round(mode(discrete_post_predictive),3))
message("From the discretization we obtain the expectation: ", round(mean(discrete_post_predictive),3))
message("From the discretization we obtain the standard deviation: ", round(sd(discrete_post_predictive),3),"\n")
```

```python
conjugate_posterior <- rbbinom(N, n, alpha0 + x, beta0 + n - x)
message("Expected value of Beta(alpha_0, beta_0) = ", round(mean(conjugate_posterior), 1))
message("Standard deviation of Beta(alpha_0, beta_0) = ", round(sd(conjugate_posterior), 1))
```

Again, very close values, which confirms our results.
