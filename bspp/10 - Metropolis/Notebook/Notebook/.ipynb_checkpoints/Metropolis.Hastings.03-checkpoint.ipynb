### Likelihood

We assume an underlying $\operatorname{N}(\mu,\sigma^2)$ distribution for the height variable, 
but what we are actually  observing is a set of pairs (intervals, counts), that is, 
multinomial data with a vector of probabilities given by the latent normal distribution. 
The likelihood is, thus,

\begin{align}
    \cal{L}(\mu,\sigma)\propto &\;\Phi(66,\mu,\sigma)^{14}
    \cdot(\Phi(68,\mu,\sigma)-\Phi(66,\mu,\sigma))^{30}\\
    &\;\cdot(\Phi(70,\mu,\sigma)-\Phi(68,\mu,\sigma))^{49}
    \cdot(\Phi(72,\mu,\sigma)-\Phi(70,\mu,\sigma))^{70}\\
    &\;\cdot(\Phi(74,\mu,\sigma)-\Phi(72,\mu,\sigma))^{33}
    \cdot(1-\Phi(68,\mu,\sigma))^{15},
\end{align}

where $\Phi(\,\cdot\,,\mu,\sigma)$ is the cdf for a normal distribution mith mean $\mu$ and 
standard deviation $\sigma$.

### Prior

We assign the usual noninformative prior pdf: the product of the improper uniform on $\mathbb{R}$ for $\mu$ and the pdf proportional to $1/\sigma$ for sigma. This improper pdf for $\sigma$ is the result of transforming back in terms of $\sigma$ the (also improper) uniform NIP on $\mathbb{R}$ for $\lambda=\log\sigma$.

$$
    h(\mu,\sigma) \propto \dfrac{1}{\sigma}.
$$

### Posterior

The posterior pdf is: 

$$
    g(\mu,\sigma\,|\,\text{data}) \propto \dfrac{1}{\sigma}\,\cal{L}(\mu,\sigma).
$$

Now we apply the parameter transformation: $\lambda=\log(\sigma)$, which yields:

$$
    g(\mu,\sigma\,|\,\text{data}) \propto \cal{L}(\mu,\exp(\lambda)).
$$

We find this posterior implemented as the `groupeddatapost` function in the `LearnBayes` package:

```
groupeddatapost<-function (theta, data){
    dj<-function(f,int.lo,int.hi,mu,sigma) f*log(pnorm(int.hi,mu,sigma)-pnorm(int.lo,mu,sigma))
    mu<-theta[1]
    sigma<-exp(theta[2])
    sum(dj(data$f,data$int.lo,data$int.hi,mu,sigma))
    }
```

### Laplace approximation

We take the (approximate) values of the ML estimates of $\mu$ and $\log(\sigma)$ as starting values in the Laplace determination of the posterior mode.

````r
#install.packages("LearnBayes",repos="https://cloud.r-project.org")
require(LearnBayes)
#
d=list(int.lo=c(-Inf,seq(66,74,by=2)),int.hi=c(seq(66,74,by=2),Inf),f=c(14,30,49,70,33,15))
#groupeddatapost<-function (theta, data){
#    dj<-function(f,int.lo,int.hi,mu,sigma) f*log(pnorm(int.hi,mu,sigma)-pnorm(int.lo,mu,sigma))
#    mu<-theta[1]
#    sigma<-exp(theta[2])
#    sum(dj(data$f,data$int.lo,data$int.hi,mu,sigma))
#    }
#
start<-c(70,1)
fit<-laplace(groupeddatapost,start,d)
fit
````

### Metropolis-Hastings

We use the output from the function `laplace` to design a Metropolis random walk algorithm to simulate from the joint posterior. For the proposal density, we use the variance-covariance matrix obtained from `laplace` and we set the scale parameter equal to 2. 

The Metropolis-Hastings algorithm is implemented in the `rwmetrop` function from the `LearnBayes` package. You should list the function and analyze each step. You will see that the actual proposal pdf is a multivariate normal (bivariate in this case) whose variance-covariance matrix is obtained by scaling the triangular matrix resulting from the Choleski decomposition of the given `var`matrix. 

We run 10,000 iterations of the random walk algorithm starting at the value start. The output `fit2` is a list with two
components: `par` is a matrix of simulated values where each row corresponds to a single draw of the parameter vector, and `accept` gives the acceptance rate of the random walk chain.

````r
proposal<-list(var=fit$var,scale=2)
fit2=rwmetrop(groupeddatapost,proposal,start,10000,d)
fit2$accept
````

### Summary statistics

From the output chain and comparing with the corresponding quantities ifrom the Laplace approximation:

````r
# from the Laplace aproximation
modal.sds=sqrt(diag(fit$var))
cbind(c(fit$mode),modal.sds)
# from the Metropolis-Hastings chain
post.means=apply(fit2$par,2,mean)
post.sds=apply(fit2$par,2,sd)
cbind(post.means,post.sds)
````

#### Chain points superimposed on a contour plot of the posterior pdf

````r
 mycontour(groupeddatapost,c(69,71,.6,1.3),d,xlab="mu",ylab="log sigma")
 points(fit2$par[5001:10000,1],fit2$par[5001:10000,2],pch=19,cex=0.01,col="blue")
````

#### What happens with a poor choice of starting point?

Suppose we rerun the Metropolis algorithm for the grouped data posterior with poor choices of starting value
and proposal density. 

As a starting value, we choose $(\mu,\log\sigma)=(65,1)$ (the choice of $\mu$ is too small) and we select 
the small scale factor of 0.2 (instead of 2):

````r
d=list(int.lo=c(-Inf,seq(66,74,by=2)),int.hi=c(seq(66,74,by=2), Inf),f=c(14,30,49,70,33,15))
start=c(70,1)
fit=laplace(groupeddatapost,start,d)
start=c(65,1)
proposal=list(var=fit$var,scale=0.2)
bayesfit=rwmetrop(groupeddatapost,proposal,start,10000,d)
bayesfit$accept
````

The acceptance rate of this modified algorithm is 0.89, which is much larger than the 0.28 rate that we found using the scale factor 2 (smaller proposed steps).

The first 2000 iterations are discarded to remove the burnIn period due to the poor starting value.

Following Albert's book, we use functions from the `coda`  and `lattice` packages to display the traceplots:

````r
require(coda)    # MCMC Convergence diagnostics
require(lattice)
dimnames(bayesfit$par)[[2]]=c("mu","log sigma")
xyplot(mcmc(bayesfit$par[-c(1:2000),]),col="black")
````

The simulated draws appear to have reached the main support of the posterior of $\mu$. However the simulated sequence appears irregular; for example, the iterates will explore the region where $\mu>70.5$ for a while before returning to the center of the distribution.

One can observe the strong correlation structure of the sequences by using autocorrelation plots. To be compared to the previously obtained with the better starting points.

````r
dimnames(bayesfit$par)[[2]]=c("mu","log sigma")
acf(bayesfit$par[-c(1:2000),])
dimnames(fit2$par)[[2]]=c("mu","log sigma")
acf(fit2$par[-c(1:2000),])
````

#### Empirical mean and standard deviation of the chains

````r
summary(mcmc(bayesfit$par[-c(1:2000),]))
batchSE(mcmc(bayesfit$par[-c(1:2000),]), batchSize=50)
````

````r
summary(mcmc(fit2$par[-c(1:2000),]))
batchSE(mcmc(fit2$par[-c(1:2000),]), batchSize=50)
````
