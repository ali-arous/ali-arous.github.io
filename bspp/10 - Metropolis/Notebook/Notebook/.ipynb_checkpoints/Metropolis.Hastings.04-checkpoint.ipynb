(2) Write a Metropolis-Hastings code with the candidate proposal kernel:

$$
	k(\theta'\,|\,\theta)=\exp\{-\frac{1}{2}(\theta'-\theta)^{2}\}.
$$	

With this candidate proposal kernel, the probability of acceptance in the Metropolis-Hastings formula is:

$$
    \min\left\{1,\frac{h_{x}(\theta')\,k(\theta\,|\,\theta')}
            {h_{x}(\theta)\,k(\theta'\,|\,\theta)}\right\}
            =\min\left\{1,\frac{h_{x}(\theta')}{h_{x}(\theta)}\right\}.
$$

The symmetry of the candidate generation implies this M-H algorithm reduces to the simple Metropolis algorithm. 

_Hint:_ Use $\theta=2$ as the start value in the trajectory. 

Test the algorithm with several values of chain length and _burn-in_ initial discarded segment.

(3) Write a Metropolis-Hastings code with the candidate proposal kernel:

$$
	k(\theta'\,|\,\theta)=q(\theta')=\exp\{-\frac{1}{2}\,\left(\theta'/3\right)^{2}\},
$$	

a $\operatorname{N}(0,3^{2})$ pdf. M-H algorithm using such a candidate proposal kernel is called _independent candidate M-H algorithm._ With it, the probability of acceptance is:
    
$$
    \min\left\{1,\frac{h_{x}(\theta')\,k(\theta\,|\,\theta')}{h_{x}(\theta)\,k(\theta'\,|\,\theta)}\right\}
		=\min\left\{1,\frac{h_{x}(\theta')\cdot q(\theta)}{h_{x}(\theta)\cdot q(\theta')}\right\}.
$$

_Hint:_ Use $\theta=0.4448$ as the start value in the trajectory. 

Test the algorithm with several values of chain length and _burn-in_ initial discarded segment.

```r
#
# Insert here your code
#
# - - - - - - - - SNIP - - - - - - - - SNIP - - - - - - - - 
gamma<-0.8
m1<-0
sd1<-1
m2<-3
sd2<-2
g<-function(theta){gamma*dnorm(theta,mean=m1,sd=sd1)+(1-gamma)*dnorm(theta,mean=m2,sd=sd2)}
u<-seq(-6,9,length=600)
plot(u,g(u),type="l",lwd=2.5,col="blue",main="Normal mixture")
```

#### (1) Generating an independent sample for this mixture pdf

```r
require(MASS)
N<-10000
Ind<-rbinom(N,size=1,prob=gamma)
theta<-Ind*rnorm(N,mean=m1,sd=sd1)+(1-Ind)*rnorm(N,mean=m2,sd=sd2)
truehist(theta,ylim=c(0,0.35),main="Normal mixture independent simulation")
lines(u,g(u),lwd=2.5,col="blue",xlab="",ylab="")
```

#### (2) Metropolis algorithm

#### Candidate proposal kernel

Here $k(t_1\,|\,t)=\exp(-(t_1-t)^2/2)$. That is: the difference $t_1-t$ follows a standard N(0,1) pdf. We add a standard deviation, in order to be able to modify the average step length along the random walk.

```r
sd.1<-1
rp.1<-function(t){rnorm(1,mean=t,sd=sd.1)}
```

#### Chain simulation

```r
metropolis<-function(start,g,rp,N.iter){
    epsilon<-1.0e-8
    chain<-rep(0,N.iter+1)
    chain[1]<-start
    for (i in 1:N.iter){
        candidate<-rp(chain[i])
        pp<-g(chain[i])
        if(pp<epsilon)
            prob<-1
        else
            prob<-min(1,g(candidate)/pp)
        if (runif(1) < prob)
            chain[i+1]<-candidate
        else
            chain[i+1]<-chain[i]
        }
    return(chain)
    }
```

#### Run the simulation

```r
N.iter<-20000
burnIn<-5000
sd.1<-2.0
start<-2.0
M.chain<-metropolis(start,g,rp.1,N.iter)
```

#### Traceplot and ACF of whole chain

```r
old.par<-par(mfrow=c(2,1))
plot(M.chain,type="l",col="cyan",ylab="theta",main="Trajectory of the whole chain")
acf(M.chain,col="cyan",main="ACF")
par(old.par)
```

#### Histogram of the chain

```r
require(MASS)
truehist(M.chain,ylim=c(0,0.35),main="Normal mixture Metropolis chain")
lines(u,g(u),lwd=2.5,col="blue",xlab="",ylab="")
```

#### Discard burnIn segment. New traceplot and ACF

```r
burnIn<-5000
M.chain.1<-M.chain[(burnIn+1):length(M.chain)]
old.par<-par(mfrow=c(2,1))
plot(M.chain.1,type="l",col="cyan",ylab="theta",main="Trajectory")
acf(M.chain.1,col="cyan",main="ACF")
par(old.par)
```

#### Histogram of the chain after discarding the `burnIn` segment

```r
require(MASS)
truehist(M.chain.1,ylim=c(0,0.35),main="Normal mixture Metropolis chain - burnIn discarded")
lines(u,g(u),lwd=2.5,col="blue",xlab="",ylab="")
```

#### (3) Metropolis-Hastings with an independent candidate generation 

A Metropolis-Hastings algorithm with a candidate proposal kernel which does not depend on the previous state $\theta$, 
that is, $k(\theta'\,|\,\theta)=q(\theta')$, is called _independent candidate_ M-H algorithm.

#### Candidate proposal kernel

```r
rq<-function(){rnorm(1,mean=0,sd=3)}
dq<-function(t){dnorm(t,mean=0,sd=3)}
```

#### M-H chain iteration

With target $g(t)$, and independent candidate density $\operatorname{dq}(t)$ and generation of random numbers $\operatorname{rq}()$.

```r
MH<-function(N.iterations,g,dq,rq,start){
    epsilon<-1.0e-10
    chain<-rep(0,N.iterations+1)
    chain[1]<-start
    for (i in 1:N.iterations){
        t<-chain[i]
        t1<-rq()
        if (g(t)*dq(t1)<epsilon)
            prob<-1
        else
            prob<-min(1,(g(t1)*dq(t))/(g(t)*dq(t1)))
        if (runif(1) < prob)
            chain[i+1]<-t1
        else
            chain[i+1]<-t
        }
    return(chain)
    }
```

#### Run the M-H chain iteration

```r
N.iterations<-20000
burnIn<-6000
start<-1.0
MH.chain<-MH(N,g,dq,rq,start)
```

#### Traceplot, ACF

```r
old.par<-par(mfrow=c(2,1))
plot(MH.chain,type="l",col="cyan",ylab="theta",main="Trajectory of the whole M-H chain")
acf(MH.chain,col="cyan",main="ACF")
par(old.par)
```

#### Histogram of the M-H chain

```r
require(MASS)
truehist(MH.chain,ylim=c(0,0.35),main="Normal mixture Metropolis-Hastings chain")
lines(u,g(u),lwd=2.5,col="blue",xlab="",ylab="")
```

#### Discard `burnIn` segment. Plot traceplot and ACF

```r
burnIn<-6000
MH.chain.1<-MH.chain[(burnIn+1):length(MH.chain)]
old.par<-par(mfrow=c(2,1))
plot(MH.chain.1,type="l",col="cyan",ylab="theta",main="Trajectory")
acf(MH.chain.1,col="cyan",main="ACF")
par(old.par)
```

#### Histogram of the chain after discarding the `burnIn` segment

```r
require(MASS)
truehist(MH.chain.1,ylim=c(0,0.35),main="Normal mixture Metropolis-Hastings chain - burnIn discarded")
lines(u,g(u),lwd=2.5,col="blue",xlab="",ylab="")
```
