Each pair $(y_j,n_j)$ represents the number $y_j$ of cancer deaths out of the the number $n_j$ of individuals at risk.

$$

Notations: 

$$
    N=\sum_{i=1}^{k}n_{i},\quad Y=\sum_{i=1}^{k}y_{i},\quad f_{j}=\dfrac{y_{j}}{n_{j}},
    \quad w_{j}=\dfrac{n_{j}}{N},\quad 1\leq j\leq k.
$$

ML estimation of the $p$, the common probability of death, assuming homogeneous Binomial populations.

$$
    \widehat{p}_{ML}=\sum_{i=1}^{k}w_{i}\cdot f_{i} = \dfrac{Y}{N}.
$$

````r
require(LearnBayes)
data(cancermortality)
y<-cancermortality$y
n<-cancermortality$n
k<-length(y)
N<-sum(n)
Y<-sum(y)
f<-y/n
w<-n/N
p.hat.ML<-sum(f*w)
k
N
Y
round(f,4)
round(w,4)
# Estimation of a common p
round(p.hat.ML,6)
````

#### Overdispersion

We compute the variance of these data: first by assuming only $k$ observations $y_{j}$ with weights $w_{j}$, next by assuming a Bernoulli or Binomial likelihood. In this second case there are two possibilities, depending on whether we assume or not a common probability for all $k$ binomials.

In any of these assumptions we find the raw variance of the observed data is much larger than what could be expected if the model were indeed Binomial. Thus we are bound to discard the Binomial model and find another discrete distribution allowing for _overdisperse data_ such as these.

````r
# Empirical expectation and variance (raw)
y.bar<-sum(w*y)
y.bar
#[1] 40.79601
s.y.2<-sum(w*(y-y.bar)^2)
round(s.y.2,2)
# Variance of observed mortality counts, assuming N Bernoulli r.v. with common p
Var.Bernoulli<-N*p.hat.ML*(1-p.hat.ML)
round(Var.Bernoulli,2)
# Variance of observed mortality counts, assuming k Binomials B(n_j,p_j), with weights w_j
Var.Binom.1<-sum(w*n*f*(1-f))
round(Var.Binom.1,2)
# Variance of observed mortality counts, assuming k Binomials B(n_j,p), with weights w_j
Var.Binom.2<-sum(w*n*p.hat.ML*(1-p.hat.ML))
round(Var.Binom.2,2)
````

#### Beta-Binomial model

The Beta-Binomial distribution, which we first met as the prior predictive distribution for a Binomial model with a Beta prior, is just _a discrete distribution_ with support on the interval of integers from 0 to $n$. Its variance can be larger than that of the Binomial, depending on an additional parameter. We propose to use it here as the likelihood for each observation $y_{i}$:

$$
	f(y\,|\,n,\alpha,\beta)=\binom{n}{y}\cdot
    \dfrac{\operatorname{B}(\alpha+y,\beta+n-y)}{\operatorname{B}(\alpha,\beta)}.
$$

where $\operatorname{B}(\,\cdot\,,\,\cdot\,)$ is a Beta function. 

Expectation and variance:
    
\begin{align*}
	\operatorname{E}(Y)&=n\cdot\dfrac{\alpha}{\alpha+\beta},\\
	\operatorname{var}(Y)&=n\cdot\dfrac{\alpha\beta}{(\alpha+\beta)^{2}}\cdot
	\dfrac{\alpha+\beta+n}{\alpha+\beta+1}.
\end{align*}


There are two usual reparameterizations of the Beta-Binomial pdf. Their goal is to have the expectation as a single parameter and, at the same time, to explicitly show expectation and variance in comparison with those of the Binomial. In both of them $\eta\in(0,1)$ has the meaning of a probability, as in the Binomial:

##### Reparameterization (1)


$$
    {\renewcommand{\arraystretch}{2.1}\begin{array}{cl}
	\eta	&=\dfrac{\alpha}{\alpha+\beta}, \mskip40mu \rho=\dfrac{1}{(\alpha+\beta+1)},
    \mskip40mu \eta\in(0,1),\quad \rho\in(0,1),\\
	\alpha	&=\eta\cdot\dfrac{1-\rho}{\rho},\quad \beta=(1-\eta)\cdot\dfrac{1-\rho}{\rho},\\
	\operatorname{E}(Y)&=n\cdot\eta,\\
	\operatorname{var}(Y)&=n\cdot\eta\cdot(1-\eta)\cdot[1+(n-1)\,\rho].
    \end{array}}
$$

##### Reparameterization (2)

$$
    {\renewcommand{\arraystretch}{2.1}\begin{array}{cl}
	K		&=\dfrac{1-\rho}{\rho}=\alpha+\beta, \mskip40mu \rho=\dfrac{1}{(K+1)},
    \mskip40mu \eta\in(0,1),\quad K\in(0,\infty),\\
	\operatorname{E}(Y)&=n\cdot\eta,\\
	\operatorname{var}(Y)&=n\cdot\eta\cdot(1-\eta)\cdot\dfrac{K+n}{K+1}.
  \end{array}}
$$

#### Beta-Binomial likelihood with the $(\eta,K)$ parameterization

$$
	f(y\,|\,n,\eta,K)=\binom{n}{y}\cdot
    \dfrac{\operatorname{B}(K\,\eta+y,K\,(1-\eta)+n-y)}{\operatorname{B}(K\,\eta,K\,(1-\eta))}.
$$


#### Prior pdf

A non-informative prior (NIP) is the following:
	
$$
	g(\eta,K)\propto \dfrac{1}{\eta\cdot(1-\eta)}\cdot \dfrac{1}{(1+K)^{2}},
    \mskip40mu 0<\eta<1, \quad 0<K<\infty,
$$

which is the product of its marginals, hence we are assuming independent parameters. The reason why the $\eta$ marginal is an NIP becomes aparent with the following reparameterization, which transforms it into an improper uniform pdf on $\mathbb{R}$. It can be checked that the $K$ marginal is the result of transforming the uniform NIP $\rho\sim\operatorname{Unif}(0,1)$.

##### Straightening out parameters

Transform $(\eta, K)$ to parameters $(\theta_{1},\theta_{2})$ defined on the whole $\mathbb{R}^2$ (as opposed to the region $[0,1]\times (0,\infty)$).

$$	
{\renewcommand{\arraystretch}{2.8}\begin{array}{lcllcl}
	\theta_{1}&=&\operatorname{logit}(\eta)=\log\left(\dfrac{\eta}{1-\eta}\right),
	&\theta_{2}&=&\log(K),\\
	\eta	&=&\dfrac{e^{\theta_{1}}}{1+e^{\theta_{1}}},
			&K	&=&e^{\theta_{2}}.
\end{array}}%
$$

##### Prior in terms of  $(\theta_{1},\theta_{2})$

Applying the change of variables for pdf's formula, we obtain the joint prior pdf in terms of the parameters $(\theta_{1},\theta_{2})$:

$$	
{\renewcommand{\arraystretch}{2.8}\begin{array}{lcll}
	g_{\theta_{1}}(\theta_{1})	&=&1,&\mskip50mu \theta_{1}\in\mathbb{R},\\
	g_{\theta_{1}}(\theta_{2})	&=&\dfrac{e^{\theta_{2}}}{(1+e^{\theta_{2}})^{2}},
    &\mskip50mu \theta_{2}\in\mathbb{R}.
\end{array}}%
$$

#### Joint posterior pdf

Up to a proportionality constant (containing both the binomial coefficient and the $y$ marginal denominator, the joint posterior pdf of $(\eta,K)$, given the observed data $\mathbf{y}=(y_1,\dots,y_n)$, is:

$$
	g(\eta,K\,|\,\mathbf{y})\propto \dfrac{1}{\eta\cdot(1-\eta)}\cdot \dfrac{1}{(1+K)^{2}}\cdot
    \prod_{i=1}^{k}\dfrac{\operatorname{B}(K\,\eta+y_i,K\,(1-\eta)+n_i-y_i)}{\operatorname{B}(K\,\eta,K\,(1-\eta))}.
$$

Now we follow closely the computations described in Chapter 5 of Jim Albert's book, implemented in the `LearnBayes` package, using such functions as `betabinexch0` (compute the posterior pdf in the $(\eta,K)$ parameterization), `mycontour` (an adaptation of the default `contour` function provided by R) `betabinexch` (compute the posterior pdf in the $(\theta_1,\theta_2)$ parameterization), and `laplace` (multivariate Laplace approximation) and code chunks `LearnBayes::Chapter.5.4` and `LearnBayes::Chapter.5.6` shown in the Code demos section.

The `betabinexch0` function implements the above expression:

```
betabinexch0<-function (theta, data){
    eta <- theta[1]
    K <- theta[2]
    y <- data[, 1]
    n <- data[, 2]
    N <- length(y)
    logf <- function(y, n, K, eta) lbeta(K * eta + y, K * (1 - 
        eta) + n - y) - lbeta(K * eta, K * (1 - eta))
    val <- sum(logf(y, n, K, eta))
    val <- val - 2 * log(1 + K) - log(eta) - log(1 - eta)
    return(val)
    }
```

The first step is to visualize this joint posterior pdf:

````r
mycontour(betabinexch0,c(.0001,.003,1,20000),cancermortality,xlab="eta",ylab="K")
````

We observe the strong skewness of the distribution, especially towards large $K$ values. This right-skewness is a common characteristic of the likelihood function of a precision ro variance parameter, with support on $(0,\infty)$.

We transform the joint posterior pdf to the new variables $(\theta_1,\theta_2)$ defined above. The result is:

$$
	g(\theta_1,\theta_2\,|\,\mathbf{y})=g(\eta(\theta_1),K(\theta_2)\,|\,\mathbf{y})\cdot
    \dfrac{e^{\theta_1+\theta_2}}{(1+e^{\theta_1})^2}
$$

#### Joint posterior pdf in terms of  $(\theta_1,\theta_2)$

The `betabinexch`implements this parameterization of the joint posterior pdf. We use it to visualize it.

```
betabinexch<-function (theta, data){
    eta<-exp(theta[1])/(1+exp(theta[1]))
    K<-exp(theta[2])
    y<-data[, 1]
    n<-data[, 2]
    N<-length(y)
    logf<-function(y,n,K,eta) lbeta(K*eta+y,K*(1-eta)+n-y)-lbeta(K*eta,K*(1-eta))
    val<-sum(logf(y,n,K,eta))
    val<- val+theta[2]-2*log(1+exp(theta[2]))
    return(val)
    }
```

#### Laplace approximation

The `laplace` function in the `LearnBayes` package implements it:

```
laplace<-function (logpost,mode,...){
    options(warn = -1)
    fit<-optim(mode, logpost, gr = NULL, ..., hessian = TRUE,control = list(fnscale = -1))
    options(warn = 0)
    mode<-fit$par
    h<- -solve(fit$hessian)
    p<-length(mode)
    int<-p/2 * log(2 * pi) + 0.5 * log(det(h)) + logpost(mode, ...)
    stuff = list(mode = mode, var = h, int = int, converge = fit$convergence == 0)
    return(stuff)
    }
```

The function optimizes the input `logpost` function, the logarithm of a joint posterior, locating its mode. Then it returns this mode together with the inverse matrix of the second derivative (Hessian) at this mode, which will become the matrix of variances and covariances of the approximating normal pdf.

````r
# The second argument to this function is an initial guess for the mode
fit=laplace(betabinexch,c(-7,6),cancermortality)
fit
````

Next we visualize the contour lines of the (logarithm of the) bivariate normal pdf with expectation equal to the `mode` and variance equal to the `var` obtained from the `laplace` function.

There are several implementations of the multivariate normal pdf in R. The most common is in the `mvtnorm` package. Here we follow Jim Albert's text, using his implementation `lbinorm` of the logarithm of the bivariate normal pdf.

````r
npar<-list(m=fit$mode,v=fit$var)
mycontour(lbinorm,c(-8,-4.5,3,16.5),npar,xlab="logit eta", ylab="log K")
````

#### Using the Laplace approximation

We take advantage of the results to obtain approximate 90% posterior credible intervals for $\theta_1=\operatorname{logit}(\eta)$ and $\theta_2=\log K$:

````r
# variances of the components of a bivariate normal vector are in the diagonal of the covariance matrix
se<-sqrt(diag(fit$var))
a<-qnorm(0.95)           # 5% probability on each tail
left.bound<-fit$mode-a*se
right.bound<-fit$mode+a*se
round(left.bound,4)
round(right.bound,4)
````

## 1. Rejection sampling

To simulate the posterior pdf $g(\theta_{1},\theta_{2}\,|\,\mathbf{y})$. Choice of candidate function $p(\,\cdot\,)$: 

- The bivariate normal  resulting from the Laplace approximation (fails: does not majorize $g$)
		
- A bivariate Student's t with the same location and scale (succeeds: it has heavier tails) 

#####  Univariate Student's t pdf

The standard univariate Student's t pdf with $\nu$  _degrees of freedom_ is:

$$ 
	f(t)={\dfrac{\Gamma((\nu +1)/2)}{\sqrt{\pi\,\nu}\cdot\Gamma (\nu/2)}}
    \times\left(1+{\frac {t^{2}}{\nu }}\right)^{\!-{(\nu +1)/2}}, 
$$

for $t\in\mathbb{R}$. A translation and scale transformation gives the univariate Student's t with $\nu$ _degrees of freedom,_ _location_ $\mu$ and _scale_ $\sigma$:

$$ 
	f(x)=\dfrac{\Gamma((\nu +1)/2)}{\sqrt{\pi\,\nu}\cdot\Gamma (\nu/2)\cdot\sigma}
    \times\left(1+\frac{1}{\nu }\,\left(\frac{x-\mu}{\sigma}\right)^2 \right)^{\!-{(\nu +1)/2}}, 
$$


##### Multivariate Student's t pdf

The $p$-dimensional Student's $t$ pdf with _(location)_ $\boldsymbol{\mu}\in\mathbb{R}^{p}$ and  _(scale)_ $\boldsymbol{\Sigma}$, a $p\times p$ positive definite (nonsingular) matrix: 

$$
    f(\mathbf{x}\,|\,\boldsymbol{\mu},\boldsymbol{\Sigma})
    =\dfrac{\Gamma((\nu+p)/2)}{(\pi\,\nu)^{p/2}\cdot\Gamma(\nu/2)\cdot(\det|\boldsymbol{\Sigma}|)^{1/2}}
   \times\left(1+\frac{1}{\nu}\, (\mathbf{x}-\boldsymbol{\mu})'\cdot\boldsymbol{\Sigma}^{-1}\cdot
   (\mathbf{x}-\boldsymbol{\mu})\right)^{-(\nu+p)/2},
$$

$\mathbf{x}\in\mathbb{R}^{p}$.

#### Rejection sampling with a candidate Student's t

We will sample from the posterior  pdf $g(\theta_{1},\theta_{2}\,|\,\mathbf{y})$. As a candidate pdf $p(\boldsymbol{\theta})$we use a multivariate t with location `fit$mode`, scale matrix `2*fit$var` (to ensure it dominates the target pdf), and 4 degrees of freedom.

We need to determine a constant $c$ such that $g(\boldsymbol{\theta}\,|\,\mathbf{y})\leq c\,p(\boldsymbol{\theta})$, for all $\boldsymbol{\theta}$. Equivalently, we need to determine a constant $d=\log c$ such that:

$$
    \log g(\boldsymbol{\theta}\,|\,\mathbf{y})-\log p(\boldsymbol{\theta})\leq d, \quad
    \text{for all  } \boldsymbol{\theta}.
$$

To do this, we harness the capabilities of the `laplace` function to obtain $d$ as the maximum of the difference:

```
betabinT<-function(theta,datapar){
    data<-datapar$data
    tpar<-datapar$par
    d<-betabinexch(theta,data)-dmt(theta,mean=c(tpar$m),
    S<-tpar$var,df=tpar$df,log=TRUE)
    return(d)
    }
```

````r
betabinT<-function(theta,datapar){
    data<-datapar$data
    tpar<-datapar$par
    d<-betabinexch(theta,data)-dmt(theta,mean=c(tpar$m),
    S<-tpar$var,df=tpar$df,log=TRUE)
    return(d)
    }
tpar<-list(m=fit$mode,var=2*fit$var,df=4)
datapar<-list(data=cancermortality,par=tpar)
start<-c(-6.9,12.4)
fit1<-laplace(betabinT,start,datapar)
dmax<-betabinT(fit1$mode,datapar)
round(fit1$mode,4)
round(dmax,4)
````

Now the `rejectsampling` function in the `LearnBayes` package implements the acceptance-rejection procedure:

```
    rejectsampling<-function (logf,tpar,dmax,n,data){
    d<-length(tpar$m)
    theta<-rmt(n, mean = c(tpar$m), S = tpar$var, df = tpar$df)
    lf<-matrix(0,c(dim(theta)[1],1))
    for (j in 1:dim(theta)[1]) 
        lf[j]<-logf(theta[j,],data)
    lg = dmt(theta, mean = c(tpar$m), S = tpar$var, df = tpar$df, log = TRUE)
    if(d == 1){
        prob<-exp(c(lf)-lg-dmax)
        return(theta[runif(n)<prob])
        }
    else{
        prob<-exp(lf-lg-dmax)
        return(theta[runif(n)< prob,])
        }
    }
```

````r
n<-10000
theta<-rejectsampling(betabinexch,tpar,dmax,n,cancermortality)
dim(theta)
````

The acceptance rate is about 23% (quite low). We can visualize the output theta by plotting these values on the contour plot of the posterior likelihood:

````r
mycontour(betabinexch,c(-8,-4.5,3,16.5),cancermortality,xlab="logit eta",ylab="log K")
points(theta[,1],theta[,2],pch=19,col="blue",cex=0.2)
````

## 2. Importance sampling

We retake the joint posterior pdf in terms of $(\theta_1,\theta_2)$ we obtained above:

$$
	g(\theta_1,\theta_2\,|\,\mathbf{y})=g(\eta(\theta_1),K(\theta_2)\,|\,\mathbf{y})
    \cdot\dfrac{e^{\theta_1+\theta_2}}{(1+e^{\theta_1})^2}
    \propto \dfrac{1}{\eta\cdot(1-\eta)}\cdot \dfrac{1}{(1+K)^{2}}\cdot
    \prod_{i=1}^{k}\dfrac{\operatorname{B}(K\,\eta+y_i,K\,(1-\eta)+n_i-y_i)}{\operatorname{B}(K\,\eta,K\,(1-\eta))}
    \cdot\dfrac{e^{\theta_1+\theta_2}}{(1+e^{\theta_1})^2}.
$$

We apply IS to compute the posterior expectation of $\theta_2=\log K$, conditional on a given value of $\theta_1=\operatorname{logit}(\eta)$, which we will take as the $\theta_1$ component of the mode of the joint posterior pdf of $(\theta_1,\theta_2)$, namely $\theta_1=-6.81979320470292$

#### Conditional posterior of $\theta_2$, given $\theta_1$

We obtain a non-normalized conditional pdf: for the moment we do not integrate the joint posterior to obtain the $\theta_1$ marginal (a daunting task, to say the least). Hence this computation reduces to fixing $\theta_1$ to its given value, discarding any factor containing only $\theta_1$ in the above expression. Later we will integrate the result to give the conditional pdf.

The non-normalized result is:

$$
	g(\theta_2,\,|\,\mathbf{y},\theta_1) \propto 
    \dfrac{K}{(1+K)^{2}}\cdot
    \prod_{i=1}^{k}\dfrac{\operatorname{B}(K\,\eta+y_i,K\,(1-\eta)+n_i-y_i)}{\operatorname{B}(K\,\eta,K\,(1-\eta))}.
$$

This expression should be understood as a function of $(\theta_1,\theta_2)$, with $K$ and $\eta$ appearing just as shorthand for $K=\exp(\theta_2)$ and $\eta=\dfrac{e^{\theta_{1}}}{1+e^{\theta_{1}}}$.

This pdf is implemented in the function `betabinexch.cond`.

````r
theta1.mode<-fit$mode[1]
round(theta1.mode,5)
betabinexch.cond<-function (log.K, data){
    eta<-exp(theta1.mode)/(1 + exp(theta1.mode))
    K<-exp(log.K)
    y<-data[, 1];n<-data[, 2];N<-length(y)
    logf<-0*log.K
    for (j in 1:length(y))
        logf<-logf + lbeta(K * eta + y[j], K * (1 -eta) + n[j] - y[j]) - lbeta(K * eta, K * (1 - eta))
        val = logf + log.K - 2 * log(1 + K)
        return(exp(val-max(val)))
    }
I=integrate(betabinexch.cond,2,16,cancermortality)
round(I$value,5)
````

#### Gaussian candidate

Our first candidate $p(\,\cdot\,)$ is a normal with mean 8 and standard deviation 2. On the left hand side plot we see
both densities, the target pdf and the candidate normal pdf. On the right side plot we see the weight function and we can verify its (presumably unbounded) growth for large values of the argument $\theta_2=\log K$.

````r
old.par<-par(mfrow=c(1,2))
curve(betabinexch.cond(x,cancermortality)/I$value,from=3,to=16,ylab="Density", xlab="log K",lwd=3, main="Densities")
curve(dnorm(x,8,2),add=TRUE)
legend("topright",legend=c("Exact","Normal"),lwd=c(3,1))
curve(betabinexch.cond(x,cancermortality)/I$value/dnorm(x,8,2),from=3,to=16, ylab="Weight",xlab="log K",
  main="Weight = g/p")
par(old.par)
````

#### Student's t candidate

Our second candidate $p(\,\cdot\,)$ is a Student's t with 2 degrees of freedom, location parameter 8 and scale 2. On the left hand side plot we see both densities, the target pdf and the candidate Student's t pdf. On the right side plot we see the weight function, which now remains bounded for large values of the argument $\theta_2=\log K$.

#### Importance sampling with the Student's t candidate

The `impsampling` function in the `LearnBayes` package implements the importance sampling procedure with a Student's t proposal function, with parameters entered as `tpar`. Its code is:

```
impsampling<-function(logf, tpar, h, n, data){
    theta<-rmt(n,mean=c(tpar$m),S=tpar$var,df=tpar$df)
    lf<-matrix(0,c(dim(theta)[1],1))
    for (j in 1:dim(theta)[1]) 
        lf[j]<-logf(theta[j,],data)
    H = lf
    for (j in 1:dim(theta)[1]) 
        H[j]<-h(theta[j,])
    lp<-dmt(theta,mean=c(tpar$m),S=tpar$var,df = tpar$df,log = TRUE)
    #
    # md is the maximum value of the difference of logs of posterior and proposal. The purpose of this line 
    # is to standardize log weights to <= 0, hence weights to <= 1.
    #
    md<-max(lf-lp) 
    wt<-exp(lf-lp-md)
    est<-sum(wt*H)/sum(wt)
    SEest<-sqrt(sum((H-est)^2*wt^2))/sum(wt)
    return(list(est=est,se=SEest,theta=theta,wt=wt))
    }
```

````r
n<-20000
tpar<-list(m=fit$mode,var=2*fit$var,df=4)
myfunc<-function(theta)
   return(theta[2])
s<-impsampling(betabinexch,tpar,myfunc,n,cancermortality)
round(cbind(s$est,s$se),4)
````

The result we obtained is that the estimate of $\theta_2=\log K$, given  $\theta_1=-6.81979$, is 7.925, with
an estimated standard error 0.0132.

## 3. Sampling Importance Resampling (SIR)

### Simple examples of SIR

The following simple, but very insightful examples of SIR, as compared to rejection sampling, are borrowed from the highly recommendable -and copious- set of lectures by [Hedibert Freitas Lopes](http://hedibert.org).

The target pdf is a standard normal distribution, $\operatorname{N}(0,1)$, with pdf:

$$
    f(x)=\dfrac{1}{\sqrt{2\,\pi}}\,e^{-\tfrac{x^2}{2}}, \mskip50mu x\in\mathbb{R}.
$$

For both rejection sampling and SIR, two proposal pdf's are used, an obviously bad candidate, a $\operatorname{Unif}(-10,10)$ pdf:

$$
    p_1(x)=\left\{
    \begin{array}{ll}
    \tfrac{1}{20},\mskip50mu &\text{if  }x\in(-10,10),\\
    0,                     &\text{otherwise.}
    \end{array}\right.
$$

and a reasonable candidate, a Student's t with $\nu=1$ degree of freedom, a.k.a. $\operatorname{Cauchy}$ distribution, with pdf:

$$
    p_2(x)=\dfrac{1}{\pi}\cdot\dfrac{1}{1+x^{2}}, \mskip50mu x\in\mathbb{R}.
$$

````r
xxx = seq(-6,6,length=1000)
y1<-dnorm(xxx)
y2<-dunif(xxx,-10,10)
y3<-dt(xxx,df=1)
plot(xxx,y1,type="l",lwd=3.0,col="blue",main="Normal target with two proposal pdf's",ylab="",xlab="x")
lines(xxx,y2,lwd=2.5,col="cyan")
lines(xxx,y3,lwd=2.5,col="magenta")
legend("topright",legend=c("Normal","Uniform","Cauchy"),lwd=3.0,col=c("blue","cyan","magenta"))
````

#### Acceptance-rejection sampling constants and theoretical acceptance rates

We need the constants:

$$
    c_1=\max_{x}\left\{\dfrac{f(x)}{p_1(x)}\right\},\mskip50mu c_2=\max_{x}\left\{\dfrac{f(x)}{p_2(x)}\right\}.
$$

The maximum of $q_1(x)=\left\{\dfrac{f(x)}{p_1(x)}\right\}$ clearly occurs at $x=0$. From the derivative of: 

$$
    q_2(x)=\left\{\dfrac{f(x)}{p_2(x)}\right\}=\sqrt{\dfrac{\pi}{2}}\cdot (1+x^2)\cdot e^{-\tfrac{x^2}{2}},
$$

we find that the maximum occurs at $x=1$. Then, the constants needed for rejection sampling are:

````r
c.1<-(1/sqrt(2*pi))*20       # dnorm(0)/dunif(0,-10,10)
c.2<-sqrt(pi/2)*2*exp(-1/2)   # dnorm(1)/dt(1,1)
round(c.1,4)
round(c.2,4)
# Theoretical acceptance rates
AR.1<-100/c.1
AR.2<-100/c.2
round(AR.1,2)
round(AR.2,2)
````

#### Rejection sampling

````r
# Rejection methods: Uniform proposal
old.par<-par(mfrow=c(1,2))
set.seed(1231354)
N      <- 10000
draw1  <- runif(N,-10,10)
accept <- dnorm(draw1)/(c.1*0.05)
u      <- runif(N)
draws1 <- draw1[u<accept]
hist(draws1,xlab="",ylab="",main="UNIFORM PROPOSAL",prob=TRUE,ylim=c(0,0.45),xlim=c(-5,5))
lines(xxx,dnorm(xxx),col=2,lwd=2)
round(length(draws1)/length(draw1),3)   # Acceptance rate

# Rejection methods: Cauchy proposal
set.seed(1231354)
N      <- 10000
draw1  <- rt(N,1)
accept <- dnorm(draw1)/(c.2*dt(draw1,1))
u      <- runif(N)
draws1 <- draw1[u<accept]
hist(draws1,xlab="",ylab="",main="CAUCHY PROPOSAL",prob=TRUE,ylim=c(0,0.45),xlim=c(-5,5))
lines(xxx,dnorm(xxx),col=2,lwd=2)
round(length(draws1)/length(draw1),3)   # Acceptance rate
par(old.par)
````

#### SIR with the normal target and the same two proposals

SIR does not require computation of the upper bound of the quotient of target and proposal pdf's

````r
N<-10000     # Number of samples from the proposal pdf
m<-2000      # Number of resamples

old.par<-par(mfrow=c(1,2))
# SIR: Uniform proposal
set.seed(1231354)
draw1 <-runif(N,-10,10)
w     <-dnorm(draw1)/0.05
ind   <-sample(1:N,size=m,replace=TRUE,prob=w)
draws1<-draw1[ind]
hist(draws1,xlab="",ylab="",main="UNIFORM PROPOSAL",prob=TRUE,ylim=c(0,0.45),xlim=c(-5,5))
lines(xxx,dnorm(xxx),col=2,lwd=2)
round(1-length(unique(draws1))/m,3) # Proportion of redraws (redundant draws)

# SIR: Cauchy proposal
set.seed(1231354)
draw1  = rt(N,1)
w      = dnorm(draw1)/dt(draw1,1)
ind    = sample(1:N,size=m,replace=TRUE,prob=w)
draws1 = draw1[ind]
hist(draws1,xlab="",ylab="",main="CAUCHY PROPOSAL",prob=TRUE,ylim=c(0,0.45),xlim=c(-5,5))
lines(xxx,dnorm(xxx),col=2,lwd=2)
round(1-length(unique(draws1))/m,3)  # Proportion of redraws (redundant draws)
par(old.par)
````

#### SIR example from Jim Albert's book

A SIR function with a multivariate Student's t proposal pdf:

```
sir<-function(logf, tpar, n, data){
    k<-length(tpar$m)
    theta<-rmt(n, mean = c(tpar$m), S = tpar$var, df = tpar$df)
    lf<-matrix(0, c(dim(theta)[1], 1))
    for (j in 1:dim(theta)[1]) 
        lf[j]<-logf(theta[j, ], data)
    lp<-dmt(theta,mean=c(tpar$m),S=tpar$var,df=tpar$df,log=TRUE)
    md<-max(lf-lp)
    wt<-exp(lf-lp-md)
    probs<-wt/sum(wt)
    indices<-sample(1:n, size = n, prob = probs, replace = TRUE)
    if (k > 1) 
        theta<-theta[indices,]
    else 
        theta<- theta[indices]
    return(theta)
    }
```
